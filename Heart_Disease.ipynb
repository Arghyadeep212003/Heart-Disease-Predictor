```python
# Diffrentiate Dependent Variable(Y) and Independent variable(X)
Y=dataset['target']
```

```python
Y
```

```python
X=dataset.drop('target',axis=1)
```

```python
X
```

```python
# Spliting data into two parts (Train and Test)
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
# Training logistic the logisticRegression model with training data
model.fit(X_train,Y_train)
```

```python
model_predict = model.predict(X_train)
```

```python
X_train
```

```python
from sklearn.metrics import accuracy_score
```

```python
training_data_accuracy=accuracy_score(model_predict,Y_train)
```

```python
training_data_accuracy
```

```python
test_model_predict = model.predict(X_test)
```

```python
testing_data_accuracy=accuracy_score(test_model_predict,Y_test)
```

```python
testing_data_accuracy
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(X_train,Y_train)
```

```python
training_data_accuracy=accuracy_score(model_predict,Y_train)
```

```python
training_data_accuracy
```

```python
test_model_predict = model.predict(X_test)
```

```python
testing_data_accuracy=accuracy_score(test_model_predict,Y_test)
```

```python
testing_data_accuracy
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(test_model_predict,Y_test))
```
